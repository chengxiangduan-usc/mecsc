```python
alt.data_transformers.disable_max_rows()
```

```python
alt.Chart(df, width=2000, height=500).mark_line().encode(
    x=alt.X('time', axis=alt.Axis(title='time'), scale=alt.Scale(zero=False)),
    y=alt.Y('skin_temp', axis=alt.Axis(title='skin_temp'), scale=alt.Scale(zero=False)),
    tooltip = [alt.Tooltip('time', format='%Y-%m-%d %H:%M:%S'),
               alt.Tooltip('skin_temp')]
    ).properties(
        title='skin_temp dist.'
    ).interactive()
```
